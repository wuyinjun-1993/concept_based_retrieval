```python
import os
topk_ls=[10,50,100,150,200,250,300]
data_path="/data2/wuyinjun"
dataset_name="crepe"

acc_metric1 = "NDCG@1:"
acc_metric2 = "NDCG@3:"
time_metric="Time taken:"
method_ls=["two", "four"]

method_to_acc1_ls = dict()
method_to_acc2_ls = dict()
method_to_time_ls = dict()

for method in method_ls:
    acc1_ls = []
    acc2_ls = []
    time_ls = []
    for val in topk_ls:
        log_file_name=os.path.join(data_path, dataset_name, "output_full_query_full_data_clustering_method_" + method + "_topk_" + str(val) + ".txt")
        with open(log_file_name, "r") as f:
            for line in f:
                if acc_metric1 in line:
                    acc1 = float(line.split(acc_metric1)[-1].strip())
                if acc_metric2 in line:
                    acc2 = float(line.split(acc_metric2)[-1].strip())
                if time_metric in line:
                    time = float(line.split(time_metric)[-1].split("s")[0].strip())
            time_ls.append(time)
            acc1_ls.append(acc1)
            acc2_ls.append(acc2)
    method_to_acc1_ls[method] = acc1_ls
    method_to_acc2_ls[method] = acc2_ls
    method_to_time_ls[method] = time_ls

    print("topk_ls:", topk_ls)
    print("acc1_ls:", acc1_ls)
    print("acc2_ls:", acc2_ls)
    print("time_ls:", time_ls)
```

```python
import os
import numpy as np
# topk_ls=[10,50,100,150,200,250,300]
# topk_ls=[20,50,80,100,150]#,500,1000,2000,5000,10000]
# topk_ls_mscoco=[200,500,1000,2000,5000,10000]
topk_ls_mscoco=[50,100,150,200,250,300]
data_path="/data2/wuyinjun"
dataset_name="mscoco_40k"

acc_metric1 = "NDCG@1:"
acc_metric2 = "NDCG@3:"
time_metric="Time taken:"
method_ls=["two", "four"]

method_to_acc1_ls_mscoco = dict()
method_to_acc2_ls_mscoco = dict()
method_to_time_ls_mscoco = dict()

random_seed_ls = [100, 200, 300]

for random_seed in random_seed_ls:

    for method in method_ls:
        acc1_ls = []
        acc2_ls = []
        time_ls = []
        for val in topk_ls_mscoco:
            if method == "four":
                log_file_name=os.path.join(data_path, dataset_name, "output_full_query_full_data_clustering_method_" + method + "_topk_" + str(val) + "_random_" + str(random_seed) + ".txt")
            else:
                log_file_name=os.path.join(data_path, dataset_name, "output_full_query_full_data_clustering_method_" + method + "_topk_" + str(val) + "_random_" + str(random_seed) + ".txt")

            with open(log_file_name, "r") as f:
                for line in f:
                    if acc_metric1 in line:
                        acc1 = float(line.split(acc_metric1)[-1].strip())
                    if acc_metric2 in line:
                        acc2 = float(line.split(acc_metric2)[-1].strip())
                    if time_metric in line:
                        time = float(line.split(time_metric)[-1].split("s")[0].strip())
                time_ls.append(time)
                acc1_ls.append(acc1)
                acc2_ls.append(acc2)
        
        if method not in method_to_acc1_ls_mscoco:
            method_to_acc1_ls_mscoco[method] = []
            method_to_acc2_ls_mscoco[method] = []
            method_to_time_ls_mscoco[method] = []
        
        method_to_acc1_ls_mscoco[method].append(np.array(acc1_ls))
        method_to_acc2_ls_mscoco[method].append(np.array(acc2_ls))
        method_to_time_ls_mscoco[method].append(np.array(time_ls))


method_to_avg_acc1_ls_mscoco = dict()
method_to_err_bar_acc1_ls_mscoco = dict()
method_to_avg_acc2_ls_mscoco = dict()
method_to_err_bar_acc2_ls_mscoco = dict()
method_to_avg_time_ls_mscoco = dict()
method_to_err_bar_time_ls_mscoco = dict()

for method in method_ls:
    acc1_ls = np.stack(method_to_acc1_ls_mscoco[method], axis=0)
    method_to_avg_acc1_ls_mscoco[method] = np.mean(acc1_ls, axis=0)
    method_to_err_bar_acc1_ls_mscoco[method] = np.std(acc1_ls, axis=0)/len(np.sqrt(random_seed_ls))
    
    acc2_ls = np.stack(method_to_acc2_ls_mscoco[method], axis=0)
    method_to_avg_acc2_ls_mscoco[method] = np.mean(acc2_ls, axis=0)
    method_to_err_bar_acc2_ls_mscoco[method] = np.std(acc2_ls, axis=0)/len(np.sqrt(random_seed_ls))
    
    time_ls = np.stack(method_to_time_ls_mscoco[method], axis=0)
    method_to_avg_time_ls_mscoco[method] = np.mean(time_ls, axis=0)
    method_to_err_bar_time_ls_mscoco[method] = np.std(time_ls, axis=0)/len(np.sqrt(random_seed_ls))
        # print("topk_ls:", topk_ls)
        # print("acc1_ls:", acc1_ls)
        # print("acc2_ls:", acc2_ls)
        # print("time_ls:", time_ls)
        
    print("acc1_ls:", method_to_avg_acc1_ls_mscoco[method])
    print("acc1_error_ls:", method_to_err_bar_acc1_ls_mscoco[method])
    print("acc2_ls:", method_to_avg_acc2_ls_mscoco[method])
    print("acc2_error_ls:", method_to_err_bar_acc2_ls_mscoco[method])
    print("time_ls:", method_to_avg_time_ls_mscoco[method])
    print("time_error_ls:", method_to_err_bar_time_ls_mscoco[method])
    
```
