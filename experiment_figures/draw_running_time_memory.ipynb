```python
import os

root_data_path="/data2/wuyinjun/"
dataset_name="crepe"

patch_count_size_ls=[1,2,3]

method_ls=["two", "four"]
acc_metric1 = "NDCG@1:"
acc_metric2 = "NDCG@3:"
time_metric="Time taken:"
cpu_memory_metric="used CPU memory::"

method_to_acc1_ls = dict()
method_to_acc2_ls = dict()
method_to_time_ls = dict()
method_to_mem_ls = dict()

base_mem = 2.7935

base_acc1 = 0.4780

base_acc2 = 0.5801

base_time = 1.13

for method in method_ls:
    acc1_ls = [base_acc1]
    acc2_ls = [base_acc2]
    time_ls = [base_time]
    mem_ls = [base_mem]
    for patch_count in patch_count_size_ls:
        file_name = os.path.join(root_data_path, dataset_name, "output_full_query_full_data_clustering_method_" + method + "_patch_count_" + str(patch_count) + ".txt")
        with open(file_name, "r") as f:
            for line in f:
                if acc_metric1 in line:
                    acc1 = float(line.split(acc_metric1)[-1].strip())
                if acc_metric2 in line:
                    acc2 = float(line.split(acc_metric2)[-1].strip())
                if time_metric in line:
                    time = float(line.split(time_metric)[-1].split("s")[0].strip())
                if cpu_memory_metric in line:
                    mem = float(line.split(cpu_memory_metric)[-1].strip())
            time_ls.append(time)
            acc1_ls.append(acc1)
            acc2_ls.append(acc2)
            mem_ls.append(mem)
            
    method_to_acc1_ls[method] = acc1_ls
    method_to_acc2_ls[method] = acc2_ls
    method_to_time_ls[method] = time_ls
    method_to_mem_ls[method] = mem_ls

    print("acc1_ls:", acc1_ls)
    print("acc2_ls:", acc2_ls)
    print("time_ls:", time_ls)
    print("mem_ls:", mem_ls)

```
